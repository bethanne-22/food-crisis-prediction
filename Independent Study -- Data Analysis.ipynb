### IMPORT GDP PER CAPITA DATA

```python
gdp_per_capita_raw = pd.read_csv("API_NY.GDP.PCAP.CD_DS2_en_csv_v2_3930492.csv")
```

```python
list_ = gdp_per_capita_raw.columns
```

```python
years = list(list_[4:])
```

```python
years
```

```python
gdp_per_capita_raw
```

```python
gdp_per_capita = gdp_per_capita_raw.melt(id_vars=['Country Name','Country Code'], value_vars=years)
```

```python
gdp_per_capita.rename(columns={'Country Name': 'country', 'Country Code':'code', 'variable':'year', 'value':'gdp_per_capita'}, inplace=True)
```

```python
gdp_per_capita.columns
```

### Create a Country Name <> Country Code Lookup

```python
country_lookup = gdp_per_capita[['code', 'country','gdp_per_capita']].groupby(['country', 'code']).mean()
```

```python
country_lookup=country_lookup.reset_index()
```

```python
country_lookup = country_lookup[['code','country']]
```

```python
country_lookup['country'].unique()
```

```python
country_lookup = country_lookup.replace('Yemen, Rep.','Yemen')
country_lookup = country_lookup.replace('Congo, Dem. Rep.','Democratic Republic of Congo')
```

```python
country_lookup['country'].unique()
```
